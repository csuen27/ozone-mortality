### Fig. 6

#### Here we share a set of demonstrative code for plotting the mortality change rate decomposition results. This Jupyter Notebook only demonstrates the plotting, while analysis code for how to obtain the result dataset is attached in the "Main Analysis" folder. Full database can be shared upon request to Dr Haitong Z Sun (zs347@cam.ac.uk) for research collaboration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('change rate decomp.csv')
data
```

```python
f = plt.figure(num=None, figsize=(20, 10), dpi=400, facecolor='w', edgecolor='k')

P = data['prov']
X1 = data['conc']
X2 = data['pop']
X3 = data['mort']
X4 = data['aging']
X5 = data['urban']
Y  = data['total']

a = 0.50

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.bar(P, X1, width=a, color='salmon', edgecolor='black')
plt.bar(P, X2, width=a, color='mistyrose', bottom=X1, edgecolor='black')
plt.bar(P, X4, width=a, color='white', bottom=X1+X2, edgecolor='black')
plt.bar(P, X3, width=a, color='silver', bottom=0, edgecolor='black')
plt.bar(P, X5, width=a, color='whitesmoke', bottom=X3, edgecolor='black')
plt.plot(P, Y, color='black', marker='o', markerfacecolor='white', markeredgewidth=1.5, linewidth=0, markersize=10)
plt.ylabel("Change in mortality attributable to multiple features (%)", fontsize=16)

plt.ylim(-200,300)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)

ax=plt.gca()
ax.spines['top'].set_linewidth(0)
ax.spines['right'].set_linewidth(0)

plt.show()
f.savefig("Fig. 6 feature contribution decomposition of change rate.pdf", bbox_inches="tight")
```
