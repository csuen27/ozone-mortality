### Fig. 4

#### Here we share a set of demonstrative code for plotting the longitudinal trend and piling-up of O${_3}$ exposure-associaetd cardiopulmonary mortalities, adding up premature deaths from COPD and cardiovascular diseases. Full database can be shared upon request to Dr Haitong Z Sun (zs347@cam.ac.uk) for research collaboration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
mort_19 = pd.read_csv('/Volumes/Database/CHNoz/conc/ozone_19_mort_china(0919).csv')

mort_19_pt1 = mort_19.groupby(['division']).sum()[['copd_25_death_urban_est','copd_25_death_rural_est']]/1000
mort_19_pt2 = mort_19.groupby(['division']).sum()[['cvd_25_death_urban_est','cvd_25_death_rural_est']]/1000
mort_19_number_list = mort_19_pt1.merge(mort_19_pt2, how = 'left', left_index = True, right_index = True)

mort_19_number_list_sum = pd.DataFrame(mort_19_number_list.sum()).T
mort_19_number_list = mort_19_number_list.append(mort_19_number_list_sum)

mort_19_number_list['year'] = 2019
mort_19_number_list.to_csv('/Volumes/Database/CHNoz/conc/mort_19_region(1112).csv')
```
