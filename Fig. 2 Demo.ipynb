### Fig. 2

#### Here we share a set of demonstrative code for plotting the longitudeinal trend of ambient O${_3}$. Full database can be shared upon request to Dr Haitong Z Sun (zs347@cam.ac.uk) for research collaboration. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

#### To track the longitudinal trend of ambient O${_3}$, we need to import the annually archived dataset from JASMIN supercomputation server, and calculate the division/area aggregated average for summarising. Here, "division" refers to teh 7 geographical administrative regions, and "area" refers to the 4 megalopolises. 

```python
conc_90 = pd.read_csv('/Volumes/Database/CHNoz/conc/ozone_90_mort_china(0919).csv')

conc_90_div_pt1 = conc_90.groupby(['division']).mean()[['MDA8h_urban_warm_90','MDA8h_rural_warm_90']]
conc_90_div_pt2 = conc_90.groupby(['division']).mean()[['MDA8h_warm_90']]
conc_90_div_pt3 = conc_90.groupby(['division']).sum()[['population_urban_90','population_rural_90']]
conc_90_div = conc_90_div_pt1.merge(conc_90_div_pt2, how = 'left', left_index = True, right_index = True).merge(conc_90_div_pt3, how = 'left', left_index = True, right_index = True)
conc_90_div['MDA8h_popwt_warm_90'] = (conc_90_div['MDA8h_urban_warm_90']*conc_90_div['population_urban_90'] + conc_90_div['MDA8h_rural_warm_90']*conc_90_div['population_rural_90'])/(conc_90_div['population_urban_90']+conc_90_div['population_rural_90'])

conc_90_division = pd.DataFrame()
conc_90_division['urban'] = conc_90_div['MDA8h_urban_warm_90']
conc_90_division['rural'] = conc_90_div['MDA8h_rural_warm_90']
conc_90_division['mean'] = conc_90_div['MDA8h_warm_90']
conc_90_division['popwt'] = conc_90_div['MDA8h_popwt_warm_90']

conc_90_a_pt1 = conc_90.groupby(['area']).mean()[['MDA8h_urban_warm_90','MDA8h_rural_warm_90']]
conc_90_a_pt2 = conc_90.groupby(['area']).mean()[['MDA8h_warm_90']]
conc_90_a_pt3 = conc_90.groupby(['area']).sum()[['population_urban_90','population_rural_90']]
conc_90_a = conc_90_a_pt1.merge(conc_90_a_pt2, how = 'left', left_index = True, right_index = True).merge(conc_90_a_pt3, how = 'left', left_index = True, right_index = True)
conc_90_a['MDA8h_popwt_warm_90'] = (conc_90_a['MDA8h_urban_warm_90']*conc_90_a['population_urban_90'] + conc_90_a['MDA8h_rural_warm_90']*conc_90_a['population_rural_90'])/(conc_90_a['population_urban_90']+conc_90_a['population_rural_90'])

conc_90_area = pd.DataFrame()
conc_90_area['urban'] = conc_90_a['MDA8h_urban_warm_90']
conc_90_area['rural'] = conc_90_a['MDA8h_rural_warm_90']
conc_90_area['mean']  = conc_90_a['MDA8h_warm_90']
conc_90_area['popwt'] = conc_90_a['MDA8h_popwt_warm_90']

conc_90_region = conc_90_division.append(conc_90_area)
conc_90_region['year'] = 1990
conc_90_region.to_csv('/Volumes/Database/CHNoz/conc/ozone_90_region(1116).csv')
```
